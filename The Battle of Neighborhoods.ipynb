# The Battle of Neighborhoods  

## The best location to open a new cafe in New York

### Business Problem

A client wants to open a cool new cafe in New York City. She is new to the city and though she loves the whole vibe of the city, she wants to know which would be the best place for her new venture. The location could be a neighborhood where there are already a lot of cafes, a buzzing street or it could be a quiet area, open to be explored. We need to provide her the best location through which she could make the most profit.

### Data

We would require the location data for New York City to proceed further with our problem statement. Data should describe the geographical location and other details like working hours, cost, ratings, address, etc of all the cafes in the city. With this data, we would be able to cluster the cafes and recommend the right location for our client to open her dream cafe.

We will be using the Foursquare location data to solve our problem. The Foursquare API allows application developers to interact with the Foursquare platform, which is is a social location service that allows users to explore the world around them. The API itself is a RESTful set of addresses to which you can send requests and recieve a response in json fromat which is easy to work with.

We shall use the explore endpoint for the API calls to obtain the required data, with the url https://api.foursquare.com/v2/venues/explore. We are interested in the section coffee near New York. The response received as json is then cleaned and formatted into a dataframe. The dataframe contains the Venues in the category Coffee, along with it's geographical co-ordinates. This dataframe is used for clustering the cafes, by which we could arrive at a solution to our problem, i.e., the right place to open the cafe.
